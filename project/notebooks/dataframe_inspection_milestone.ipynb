## Section 1: Inspecting Data with head()

### What is head()?

`head()` displays the **first few rows** of a DataFrame.

**Purpose:**
- Quick visual preview
- See column names in context
- Check sample values
- Verify data alignment

**Default:** Shows first 5 rows  
**Custom:** Use `head(n)` for n rows

```python
# Example 1: Default head() - First 5 rows
print("First 5 rows of the dataset:")
print()
df.head()
```

```python
# Example 2: head(10) - First 10 rows
print("First 10 rows of the dataset:")
print()
df.head(10)
```

```python
# Example 3: head(3) - First 3 rows
print("First 3 rows only:")
print()
df.head(3)
```

```python
# Bonus: tail() - Last rows
print("Last 5 rows of the dataset:")
print()
df.tail()
```

### What head() Tells You:

✅ **Column names** - What variables exist?  
✅ **Sample values** - What does the data actually look like?  
✅ **Data format** - Numbers, text, dates?  
✅ **Missing values** - Any NaN values visible?  
✅ **Data alignment** - Values in correct columns?  

**IMPORTANT:** head() shows **SAMPLE** data, not ALL data. Don't make conclusions about the entire dataset from head() alone!

## Section 2: Inspecting Structure with info()

### What is info()?

`info()` provides a **concise summary** of DataFrame structure.

**Information Provided:**
- Total rows and columns
- Column names
- Non-null counts (identifies missing values)
- Data types (dtype)
- Memory usage

**This is your "health check" for the DataFrame.**

```python
# Execute info() on our dataset
print("DataFrame Structure:")
print()
df.info()
```

### Understanding info() Output:

**1. RangeIndex:** Shows the index range
- "RangeIndex: 20 entries, 0 to 19" → 20 rows

**2. Data Columns:** Lists each column
- Column name
- **Non-Null Count:** How many values are NOT missing
- **Dtype:** The data type

**3. Dtypes Summary:** Count of each data type
- `int64`: Integer numbers
- `float64`: Decimal numbers
- `object`: Text/strings or mixed
- `bool`: Boolean (True/False)

**4. Memory Usage:** RAM consumed by DataFrame

### 🔑 KEY INSIGHT:
**If Non-Null Count < Total Rows → Missing values exist!**

```python
# Let's identify columns with missing values
print("Missing Value Analysis:")
print(f"Total rows: {len(df)}")
print()
print("Per-column missing counts:")

for col in df.columns:
    non_null = df[col].notna().sum()
    missing = len(df) - non_null
    if missing > 0:
        print(f"  {col}: {missing} missing values ({missing/len(df)*100:.1f}%)")
    else:
        print(f"  {col}: Complete (0 missing)")
```

### Why info() is Critical:

✅ **Reveals data types** - Ensures columns are correct type  
✅ **Shows missing values** - Identifies data quality issues  
✅ **Memory usage** - Important for large datasets  
✅ **Quick overview** - Fast way to understand structure  

### Common Issues Caught by info():
- ❌ Numbers stored as 'object' (string) instead of int/float
- ❌ Date columns stored as 'object' instead of datetime
- ❌ Missing values you didn't know about
- ❌ Wrong number of columns after loading

## Section 3: Summarizing Data with describe()

### What is describe()?

`describe()` provides **statistical summaries** of numeric columns.

**Statistics Provided:**
- **count:** Number of non-missing values
- **mean:** Average value
- **std:** Standard deviation (spread)
- **min:** Minimum value
- **25%:** First quartile (25th percentile)
- **50%:** Median (50th percentile)
- **75%:** Third quartile (75th percentile)
- **max:** Maximum value

```python
# Execute describe() on our dataset
print("Statistical Summary:")
print()
df.describe()
```

### Understanding describe() Output:

**count:** How many non-missing values
- If count < total rows → missing values exist

**mean:** Average value
- Center of distribution
- Sensitive to outliers

**std:** Standard deviation
- How spread out values are
- Low std = values similar
- High std = values vary widely

**min/max:** Range boundaries
- Check for impossible values (e.g., negative age)
- Identify potential outliers

**25%, 50%, 75%:** Quartiles
- 50% is the median (middle value)
- Less sensitive to outliers than mean

### Distribution Clues:
- If `mean ≈ 50%` → Symmetric distribution
- If `mean >> 50%` → Right-skewed (outliers pull mean up)
- If `mean << 50%` → Left-skewed (outliers pull mean down)

```python
# Let's interpret a specific column
print("Age Column Analysis:")
print()
age_stats = df['age'].describe()
print(age_stats)
print()
print("Interpretation:")
print(f"  • Average age: {age_stats['mean']:.1f} years")
print(f"  • Age range: {age_stats['min']:.0f} to {age_stats['max']:.0f}")
print(f"  • Middle 50% between {age_stats['25%']:.0f} and {age_stats['75%']:.0f} years")
print(f"  • Median (50%): {age_stats['50%']:.0f} years")
```

```python
# Score column analysis
print("Score Column Analysis:")
print()
score_stats = df['score'].describe()
print(score_stats)
print()
print("Interpretation:")
print(f"  • Average score: {score_stats['mean']:.1f}")
print(f"  • Score range: {score_stats['min']:.0f} to {score_stats['max']:.0f}")
print(f"  • Middle 50% scored between {score_stats['25%']:.0f} and {score_stats['75%']:.0f}")
```

```python
# Include all columns (not just numeric)
print("Complete Summary (All Columns):")
print()
df.describe(include='all')
```

### Why describe() is Essential:

✅ **Quick statistics** - Understand distributions instantly  
✅ **Spot outliers** - Extreme min/max values stand out  
✅ **Check ranges** - Verify values are reasonable  
✅ **Detect issues** - Impossible values become obvious  
✅ **Plan analysis** - Know what you're working with  

### Common Issues Caught by describe():
- ❌ Negative values where they shouldn't exist (age, price)
- ❌ Extreme outliers (max >> typical values)
- ❌ Zero variance (std = 0, all values same)
- ❌ Missing values (count < total rows)

## Section 4: When to Use Each Method

### Method Comparison

| Method | Purpose | When to Use |
|--------|---------|-------------|
| `head()` | Visual preview | First look, check alignment |
| `info()` | Structural summary | Check types, find missing values |
| `describe()` | Numeric statistics | Understand distributions, spot outliers |

### The Three Essential Questions:

1. **What does it look like?** → Use `head()`
2. **How is it structured?** → Use `info()`
3. **What do the numbers mean?** → Use `describe()`

### Recommended Inspection Workflow:

```python
# Complete inspection workflow
print("="*60)
print("COMPLETE DATAFRAME INSPECTION WORKFLOW")
print("="*60)

print("\nSTEP 1: Visual Preview with head()")
print("-"*60)
print(df.head(3))

print("\n\nSTEP 2: Structural Check with info()")
print("-"*60)
df.info()

print("\n\nSTEP 3: Statistical Summary with describe()")
print("-"*60)
print(df.describe())

print("\n" + "="*60)
print("INSPECTION COMPLETE - Ready for Analysis!")
print("="*60)
```

## Section 5: Best Practices

### ✅ DO:
- **Always inspect after loading** - Make it automatic
- **Use all three methods** - Each reveals different information
- **Interpret the output** - Don't just run commands
- **Check for anomalies** - Look for weird values
- **Document findings** - Note issues for cleaning

### ❌ DON'T:
- **Skip inspection** - Never analyze without inspecting
- **Use only head()** - It's not enough alone
- **Ignore data types** - Wrong types cause errors
- **Overlook missing values** - They affect analysis
- **Forget to check the tail** - Issues hide at the end
