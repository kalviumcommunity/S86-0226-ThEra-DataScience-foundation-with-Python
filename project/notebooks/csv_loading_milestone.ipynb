## Section 1: Understanding CSV Files

**CSV** stands for **Comma-Separated Values**. It's a simple text file format for storing tabular data.

### Structure:
- First row typically contains column headers
- Each subsequent row contains data values
- Values are separated by commas (or other delimiters)
- Similar to spreadsheet tables

### Example CSV:
```
id,name,age
1,Alice,25
2,Bob,30
3,Charlie,35
```

## Section 2: Loading CSV Files

Let's import pandas and load our CSV files.

````python
# Import pandas library
import pandas as pd
import os

print("Pandas imported successfully!")
print(f"Pandas version: {pd.__version__}")
````

````python
# Define file paths
raw_data_path = '../data/raw/example_raw.csv'
processed_data_path = '../data/processed/example_processed.csv'

# Check if files exist
print("Checking file existence:")
print(f"Raw data exists: {os.path.exists(raw_data_path)}")
print(f"Processed data exists: {os.path.exists(processed_data_path)}")
````

````python
# Load the raw CSV file
# pd.read_csv() is the primary method for loading CSV files
df_raw = pd.read_csv(raw_data_path)

print("✓ Raw data loaded successfully!")
````

````python
# Load the processed CSV file
df_processed = pd.read_csv(processed_data_path)

print("✓ Processed data loaded successfully!")
````

## Section 3: Inspecting Loaded Data

**ALWAYS inspect data after loading!** This verifies the data loaded correctly and helps catch issues early.

### Key Inspection Methods:

- `df.shape` - Get dimensions (rows, columns)

- `df.describe()` - Statistical summary

## Section 4: Common Loading Issues

### Issues to Watch For:

1. **Incorrect File Path**
   - Symptom: `FileNotFoundError`
   - Solution: Verify path, use absolute paths if needed

2. **Missing Headers**
   - Symptom: First data row becomes column names
   - Solution: Use `header=None` or specify header row

3. **Wrong Delimiter**
   - Symptom: All data in one column
   - Solution: Use `sep=';'` or `delimiter` parameter

4. **Encoding Issues**
   - Symptom: Special characters appear as gibberish
   - Solution: Try `encoding='utf-8'` or `'latin-1'`

5. **Mixed Data Types**
   - Symptom: Numbers stored as strings
   - Solution: Specify `dtype` or convert after loading

### Why Inspection Matters:
- ✅ Catches errors early
- ✅ Prevents incorrect analysis
- ✅ Saves debugging time
- ✅ Ensures data integrity

## Section 5: Advanced Loading Techniques

Here are some useful parameters for `pd.read_csv()`:

````python
# Load specific columns only
df_subset = pd.read_csv(raw_data_path, usecols=['id', 'value'])

print("Loaded specific columns:")
print(f"Columns: {list(df_subset.columns)}")
df_subset
````

````python
# Load with specified data types
df_typed = pd.read_csv(raw_data_path, dtype={'id': int, 'value': int})

print("Data types specified during load:")
df_typed.dtypes
````

````python
# Load first N rows only (useful for large files)
df_limited = pd.read_csv(raw_data_path, nrows=2)

print("First 2 rows only:")
df_limited
````

## Verification Checklist

After loading any CSV file, always verify:

- ☑️ File loaded without errors
- ☑️ Number of rows matches expectations
- ☑️ Number of columns matches expectations
- ☑️ Column names are correct (no typos, no extra spaces)
- ☑️ Data types are appropriate (numbers as int/float, not strings)
- ☑️ First few rows look reasonable
- ☑️ Last few rows look reasonable
- ☑️ No unexpected missing values
- ☑️ Shape makes sense
- ☑️ Info shows expected structure
